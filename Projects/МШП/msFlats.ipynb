```python
import pandas as pd
df = pd.read_csv('Flats.csv', sep=';')
df['price'] = df['price'].apply(lambda x: float(str(x).replace(',', '.')))
df['totsp'] = df['totsp'].apply(lambda x: float(str(x).replace(',', '.')))
df['livesp'] = df['livesp'].apply(lambda x: float(str(x).replace(',', '.')))
df['kitsp'] = df['kitsp'].apply(lambda x: float(str(x).replace(',', '.')))
df['dist'] = df['dist'].apply(lambda x: float(str(x).replace(',', '.')))
df['metrdist'] = df['metrdist'].apply(lambda x: float(str(x).replace(',', '.')))
df.head()
```
