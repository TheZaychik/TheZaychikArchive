```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([9,12,12,15,16])
y = np.array([1,1,2,4,3])
plt.scatter(x,y, color='red')
plt.show()
```

```python
x = np.array([9,12,12,15,16])
y = np.array([1,1,2,4,3])
x.mean(), y.mean()
```

```python
x1 = x - x.mean()
y1 = y - y.mean()
b = sum(x1 * y1) / sum(x1*x1)
a = y.mean() - x.mean()*b
a, b
```

```python
x_plot = np.arange(8,18)
y_plot = a + b * x_plot
plt.scatter(x,y, color='red')
plt.plot(x_plot, y_plot, '--')
plt.show()
```

```python
from sklearn import linear_model as lm
regr = lm.LinearRegression()
x = x.reshape(5,1)
y = y.reshape(5,1)
regr.fit(x,y)
plt.scatter(x,y, color='red')
plt.plot(x, regr.predict(x))
plt.show()
```

```python
regr.coef_ # b
```

```python
regr.intercept_ # a                  y = a + bx
```

```python
x = x.reshape(6,1)
y = y.reshape(6,1)
regr.fit(x,y)
plt.scatter(x,y, color='red')
plt.plot(x, regr.predict(x))
plt.show()
```

```python
regr.intercept_, regr.coef_ # a, b
```

```python
x1 = x - x.mean()
y1 = y - y.mean()
b = sum(x1 * y1) / sum(x1*x1)
a = y.mean() - x.mean()*b
y_s_krishkoy = a + b * x
e = y - y_s_krishkoy
e
```

```python
r2 = 1 - sum(e**2)/sum(y1**2)
r2[0]
```

```python
from sklearn.metrics import r2_score as r2
r2(y, regr.predict(x)) # Точность расчета велика
```
