```python
import pandas as pd
```

```python
import json
with open('fifa_data.json') as f:
    data = json.loads(f.read())
```

```python
data_pandas = pd.DataFrame.from_dict(data['matches'])
```

```python
data_pandas['attendance'].astype(int).sum()
```

```python
# Добавить столбец количество событий
data_pandas['away_team_events_num'] = data_pandas['away_team_events'].apply(lambda x: len(x))
data_pandas.head()
```

```python
# Только матчи сборной России
df = data_pandas
df = df.loc[(df['away_team_country'] == 'Russia') | (df['home_team_country'] == 'Russia')]
df
```
