```python
import pandas as pd
import json
with open('fifa_data.json') as f:
    data = json.loads(f.read())
```

```python
df = pd.DataFrame.from_dict(data['matches'])
df
```

```python
df['attendance'].astype(int).mean()
```

```python
t = pd.DataFrame()
t['home_team'] =  df['home_team_country']
t['home_sub'] = df['home_team_events'].apply(lambda x: len(list(filter(lambda y:y['type_of_event'].startswith('substitution'), x))))
t['away_team'] = df['away_team_country']
t['away_sub'] = df['away_team_events'].apply(lambda x: len(list(filter(lambda y:y['type_of_event'].startswith('substitution'), x))))
t = t.groupby('home_team')['home_sub'].sum() + t.groupby('away_team')['away_sub'].sum()
t.sort_values()
```

```python
t = pd.DataFrame()
t['home_team'] =  df['home_team_country']
t['home_sub'] = df['home_team_events'].apply(lambda x: len(list(filter(lambda y:y['type_of_event'].startswith('substitution'), x))))
t['away_team'] = df['away_team_country']
t['away_sub'] = df['away_team_events'].apply(lambda x: len(list(filter(lambda y:y['type_of_event'].startswith('substitution'), x))))
t = t.groupby('home_team')['home_sub'].mean() + t.groupby('away_team')['away_sub'].mean()
t.value_counts()
```

```python
df['attendance_int'] = df['attendance'].astype(int)
df.groupby('location')['attendance_int'].sum().sort_values()
```

```python
df['matches_teams'] = df['home_team_country'] + '_' + df['away_team_country']
df.groupby('matches_teams')['attendance_int'].sum().sort_values()
```

```python
df['temp_cel'] = df['weather'].apply(lambda x:x['temp_celsius']).astype(float)
df.groupby('location')['temp_cel'].mean().sort_values()
```
